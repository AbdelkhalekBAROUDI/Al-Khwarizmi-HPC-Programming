## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Linear convection without CFL

```python
def linearconv_withoutCFL(u, nx, nt, dx):
    dt = .025  
    c = 1.
    un = np.zeros(nx)
    
    for n in range(nt):
        for i in range(nx): un[i] = u[i]
        for i in range(1, nx):
            u[i] = un[i] - c * (dt / dx) * (un[i] - un[i-1])
    return 0
```

## Linear convection with CFL

```python
def linearconv_withCFL(u, nx, nt, dx):
    c = 1.
    CFL = 0.8
    dt = CFL*dx/np.fabs(c) 
    un = np.zeros(nx)
    
    for n in range(nt):
        for i in range(nx): un[i] = u[i]
        for i in range(1, nx):
            u[i] = un[i] - c * (dt / dx) * (un[i] - un[i-1])        
    return 0
```

## Inputs

```python
nx = 93 #change to 83, 85, 91, ... what do you remark ?
dx = 2 / (nx - 1)
nt = 20    

grid = np.linspace(0,2,nx) 
u0 = np.ones(nx)      
u0[int(.5/dx):int(1 / dx + 1)] = 2  
```

## Call Funtions

### Function without CFL

```python
u = u0.copy()
linearconv_withoutCFL(u, nx, nt, dx)

plt.plot(grid,u0);
plt.plot(grid,u);
plt.ylim([1.,2.2])
plt.xlabel('X')
plt.ylabel('Velocity')    
plt.title('1D Linear Convection')  
```

We notice that the computational solution blew up.

### Function with CFL

```python
u = u0.copy()
linearconv_withCFL(u, nx, nt, dx)

plt.plot(grid,u0);
plt.plot(grid,u);
plt.ylim([1.,2.2])
plt.xlabel('X')
plt.ylabel('Velocity')    
plt.title('1D Linear Convection') 
```

We can note that after we respected the Courant–Friedrichs–Lewy condition, we ensured the stability of the discreted schema.
